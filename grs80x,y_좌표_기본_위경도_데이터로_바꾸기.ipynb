```python
from pyproj import Proj
from pyproj import transform
import pandas as pd
import numpy as np
import openpyxl

WGS84 = { 'proj':'latlong', 'datum':'WGS84', 'ellps':'WGS84', }

GRS80 = { 'proj':'tmerc', 'lat_0':'38', 'lon_0':'127', 'k':1, 'x_0':200000,
         'y_0':600000, 'ellps':'GRS80', 'units':'m' }

def grs80_to_wgs84(x, y):
  return transform( Proj(**GRS80), Proj(**WGS84), x, y)
```

```python
df = pd.read_excel('VDS 설치정보 조회 서비스_빈칸삭제.xlsx')
```

```python
df.head(3)
```

```python
newdf = df[['grs80x','grs80y']]
newdf.head(3)
```

```python
coord = np.array(newdf['grs80x'])
coord
```

```python
coord2 = np.array(newdf['grs80y'])
coord2
```

```python
cc =grs80_to_wgs84(coord,coord2)
```

```python
xcode = cc[0].tolist()
ycode = cc[1].tolist()
```

```python
binlist = []

for i, j in zip(xcode,ycode):
  tmp = i, j

  binlist.append(tmp)

  print(tmp)
```

```python
df['좌표변환'] = binlist
df.head(3)
```

```python
df.to_excel('VDS 설치정보 조회 서비스_최종.xlsx',index=False)
```
